```python
from langchain.chat_models import init_chat_model
```

```python
import easyocr
reader = easyocr.Reader(['hi', 'en'])
result = reader.readtext("../Images/Output/out0.jpg", detail = 0)
```

```python
result
```

```python
scannedText = " ".join(result)
```

```python
model = init_chat_model("gemini-2.0-flash", model_provider="google_genai")
```

```python

from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate([
    ("system", """ 
        You are a translator
        Translate the text in the given language to English
        Dont Add additional text. Only the translated text
        You will be given a document extract every detail in a well formatted JSON
    """),
    ("user", "Translate {text}")
])
prompt = prompt_template.invoke({
    "text" : scannedText
})
response = model.invoke(prompt)
response.content
```

```python
with open("../textFiles/EBHindi", "w", encoding="utf-8") as f:
    f.write(response.content)
```
